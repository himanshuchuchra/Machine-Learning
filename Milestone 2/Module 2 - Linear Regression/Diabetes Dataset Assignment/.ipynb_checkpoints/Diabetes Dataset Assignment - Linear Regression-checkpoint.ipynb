# Diabetes Dataset Assignment | Linear Regression

```python
import pandas as pd
import numpy as np
```

## Importing the Data

```python
diabetes = np.genfromtxt("diabetes_train_data.csv", dtype = None, delimiter = ",")
print(diabetes)
print(diabetes.shape)
```

## Converting Data to DataFrame

#### I am converting the data to DataFrame as the Output data is also added in the dataset and we need to separate the output data.

```python
diabetes_dataframe = pd.DataFrame(diabetes)
diabetes_dataframe
```

## Separating the Input and Output

```python
diabetes_input = diabetes_dataframe.iloc[:,0:10]
diabetes_output = diabetes_dataframe.iloc[:,10]
```

```python
diabetes_input
```

```python
diabetes_output
```

## Converting the Data to Numpy Array

```python
X_train = diabetes_input.to_numpy()
Y_train = diabetes_output.to_numpy()
X_train, Y_train
```

## Training the Linear Regression Algorithm

```python
from sklearn.linear_model import LinearRegression
alg = LinearRegression()
```

## Fit the Data in Algorithm

```python
alg.fit(X_train,Y_train)
```

## Importing the Test Data for Prediction

```python
X_test = np.genfromtxt("diabetes_test_data.csv", dtype = None, delimiter = ',')
X_test
```

## Predicting the Output on X_test Data

```python
Y_pred = alg.predict(X_test)
Y_pred
```

## Setting Predicted Output to 5 Decimal Places

```python
Y_pred = np.around(Y_pred,5)
Y_pred
```
