# Boston Dataset

```python
from sklearn import datasets
```

## Importing Boston Data

```python
boston = datasets.load_boston()
print(boston)
```

## Analysing the Data

```python
data = boston.data
target = boston.target
print(data)
print(target)
```

## Converting Data to DataFrame

```python
import pandas as pd
boston_dataframe = pd.DataFrame(data)
boston_dataframe
```

## Adding Heading to DataFrame

```python
boston_dataframe.columns = boston.feature_names
temp = None
temp_dataframe = pd.DataFrame(temp)
check = []
for i in boston_dataframe:
    name = i + '_' + i
    check = check + [name]
    boston_dataframe[i + '_' + i] = boston_dataframe[i] ** 2
    
for i in boston_dataframe:
    check = check + [i]
    for j in boston_dataframe:
        if j in check:
            continue
        else:
            temp_dataframe[i + '_' + j] = boston_dataframe[i]*boston_dataframe[j]
temp_dataframe
```

## Combining two Data Frames

```python
boston_dataframe
```

```python
for i in temp_dataframe:
    boston_dataframe[i] = temp_dataframe[i]
```

```python
boston_dataframe
```

```python
X2 = boston_dataframe.values
print(X2)
boston_dataframe.describe()
```

## Spliting the Data to Train and Test

```python
from sklearn import model_selection
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(data,target,random_state = 0)
X2_train,X2_test,Y2_train,Y2_test = model_selection.train_test_split(X2,target,random_state = 0)
```

## Import LinearRegression for Training Algorithm

```python
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg2 = LinearRegression()
```

## Fit the Data in Algorithm

```python
alg1.fit(X_train,Y_train)
alg2.fit(X2_train,Y2_train)
```

## Predict the Output

```python
X_pred = alg1.predict(X_train)
Y_pred = alg1.predict(X_test)
X2_pred = alg2.predict(X2_train)
Y2_pred = alg2.predict(X2_test)
X_pred,Y_pred,X2_pred,Y2_pred
```

## Comparing Scores

```python
def cod_score(Y_Test, Y_Pred):
    u = ((Y_Test - Y_Pred)**2).sum()
    v = ((Y_Test - Y_Test.mean())**2).sum()
    result = 1 - (u/v)
    return result
train_score = cod_score(Y_train,X_pred)
test_score = cod_score(Y_test,Y_pred)
print("Train: ",train_score)
print("Test: ",test_score)

train_score_2 = cod_score(Y2_train,X2_pred)
test_score_2 = cod_score(Y2_test,Y2_pred)
print("Train_2: ",train_score_2)
print("Test_2: ",test_score_2)
```
