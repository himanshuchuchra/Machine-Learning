```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
```

## Importing Boston Dataset

```python
boston = datasets.load_boston()
boston
```

## Initializing Data and Target Values

```python
data = boston.data
target = boston.target
data, target
```

## Converting Data to DataFrame

```python
boston_dataframe = pd.DataFrame(data)
boston_dataframe
```

#### We have converted the data to dataframe as we have to chek any null values or string in the data. If any one of them is present then we have to perform Data Cleaning.

## Checking the Data

```python
boston_dataframe.describe()
```

## Adding Combinations in Data

### So, we will be adding three different combinations and checking results according to that. Here is the list of how we are applying combinations:
1. Square of all Attributes
2. Cross Multiplication of all Attributes
3. Square + Cross Multiplication of all Attributes

### Square of All Attribures

```python
boston_dataframe_square = boston_dataframe.copy()
size = boston_dataframe_square.shape[1]
for i in boston_dataframe_square:
    name = i + size
    boston_dataframe_square[name] = boston_dataframe_square[i] ** 2
boston_dataframe_square
```

### Cross Multiplication of all Attributes

```python
boston_dataframe_cross = boston_dataframe.copy()
size = boston_dataframe_cross.shape[1]
temp = None
temp_dataframe = pd.DataFrame(temp)
check = []
count = 0
for i in boston_dataframe_cross:
    check = check + [i]
    for j in boston_dataframe_cross:
        if j in check:
            continue
        elif i == j:
            continue
        else:
            name = count + size
            temp_dataframe[name] = boston_dataframe_cross[i]*boston_dataframe_cross[j]
            count = count + 1

for i in temp_dataframe:
    boston_dataframe_cross[i] = temp_dataframe[i]
boston_dataframe_cross
```

### Square + Cross of all Attributes

```python
square_cross = None
boston_dataframe_square_cross = pd.DataFrame(square_cross)

for i in boston_dataframe_square:
    boston_dataframe_square_cross[i] = boston_dataframe_square[i]

size = boston_dataframe_square_cross.shape[1]
    
for i in temp_dataframe:
    boston_dataframe_square_cross[size] = temp_dataframe[i]
    size = size + 1
    
boston_dataframe_square_cross
```

## Converting all Combination DataFrames into Numpy Arrays

```python
square_data = boston_dataframe_square.to_numpy()
cross_data = boston_dataframe_cross.to_numpy()
square_cross_data = boston_dataframe_square_cross.to_numpy()
square_data,cross_data,square_cross_data
```

## Spliting the Data to Train and Test

```python
X1_train,X1_test,Y1_train,Y1_test = model_selection.train_test_split(data,target,random_state = 0)
X2_train,X2_test,Y2_train,Y2_test = model_selection.train_test_split(square_data,target,random_state = 0)
X3_train,X3_test,Y3_train,Y3_test = model_selection.train_test_split(cross_data,target,random_state = 0)
X4_train,X4_test,Y4_train,Y4_test = model_selection.train_test_split(square_cross_data,target,random_state = 0)
```

## Training Algorithm using LinearRegression

```python
alg1 = LinearRegression()
alg2 = LinearRegression()
alg3 = LinearRegression()
alg4 = LinearRegression()
```

### Fit Function

```python
alg1.fit(X1_train, Y1_train)
alg2.fit(X2_train, Y2_train)
alg3.fit(X3_train, Y3_train)
alg4.fit(X4_train, Y4_train)
```

### Predict Function

```python
X1_pred = alg1.predict(X1_train)
Y1_pred = alg1.predict(X1_test)

X2_pred = alg2.predict(X2_train)
Y2_pred = alg2.predict(X2_test)

X3_pred = alg3.predict(X3_train)
Y3_pred = alg3.predict(X3_test)

X4_pred = alg4.predict(X4_train)
Y4_pred = alg4.predict(X4_test)
```

## Implementing Coefficient of Determination

```python
def cod_score(Y_Test, Y_Pred):
    u = ((Y_Test - Y_Pred)**2).sum()
    v = ((Y_Test - Y_Test.mean())**2).sum()
    result = 1 - (u/v)
    return result

train_score_1 = cod_score(Y1_train,X1_pred)
test_score_1 = cod_score(Y1_test,Y1_pred)
print("Train_1: ",train_score_1)
print("Test_1: ",test_score_1)

train_score_2 = cod_score(Y2_train,X2_pred)
test_score_2 = cod_score(Y2_test,Y2_pred)
print("Train_2: ",train_score_2)
print("Test_2: ",test_score_2)

train_score_3 = cod_score(Y3_train,X3_pred)
test_score_3 = cod_score(Y3_test,Y3_pred)
print("Train_3: ",train_score_3)
print("Test_3: ",test_score_3)

train_score_4 = cod_score(Y4_train,X4_pred)
test_score_4 = cod_score(Y4_test,Y4_pred)
print("Train_4: ",train_score_4)
print("Test_4: ",test_score_4)
```
