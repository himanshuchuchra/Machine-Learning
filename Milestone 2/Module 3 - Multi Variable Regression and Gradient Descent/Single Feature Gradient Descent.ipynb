## step_gradient Function

```python
def step_gradient(data,learning_rate,m,c):
    m_slope = 0
    c_slope = 0
    N = len(data)
    for i in range(N):
        X = data[i,0]
        Y = data[i,1]
        m_slope += (-2 / N) * (Y - (m*X) - c) * (X)
        c_slope += (-2 / N) * (Y - (m*X) - c)
    new_m = m - (learning_rate * m_slope)
    new_c = c - (learning_rate * c_slope)
    return new_m,new_c
```

```python
def cost(data,m,c):
    total_cost = 0
    N = len(data)
    for i in range(N):
        X = data[i,0]
        Y = data[i,1]
        total_cost += (1/N)*((Y - m*N - c)**2)
    return total_cost
```
